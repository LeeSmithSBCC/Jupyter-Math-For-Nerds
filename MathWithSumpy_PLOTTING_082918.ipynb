```python
from sympy import *
init_printing()
from matplotlib import style
#style.use('grayscale')
```

```python
x, y, z = symbols("x, y, z")
```

```python
a, b,c,d = symbols ('a b c d')
```

```python
plot(x,x**2,x**3, xlim=(-4,4),ylim=(-4,4))
```

```python
plot(x**2+2*x+1,x**2+2*x+2,x**2+2*x+3,  xlim=(-4,4),ylim=(-4,4))
```

```python
plot(sin(x), cos(x), tan(x), (x, -4, 4),ylim=(-4,4))
```

```python
p1 = sin(x)
p2 = cos(x)
p3 = tan(x)
leePlot = plot(p1,p2,p3,xlim=(-6.5,6.5),ylim=(-4,4), show=false)
leePlot[1].line_color = 'r'
leePlot[2].line_color = 'g'
leePlot.show()
```

```python
p1=x**2+2*x+1
p2=x**2+2*x+2
p3=x**2+2*x+3
#leePlot2 = plot(p1, xlim=(-4,4),ylim=(-4,4))
leePlot2 = plot(p1,p2,p3, xlim=(-4,4),ylim=(-4,4),show=false)
leePlot2[1].line_color='r'
leePlot2[2].line_color='g'
leePlot2.grid='True'
leePlot2.show()
```

```python
p1 = (x+1)**2
p2 = (x)**2
p3 = (x-1)**2
p4 = (x-1)
leePlot3 = plot(p1, p2, p3, p4, xlim=(-4,4),ylim=(-4,4), show=false)
leePlot3[1].line_color='r'
leePlot3[2].line_color='g'
leePlot3[3].line_color='y'
leePlot3.show()
```
