```python
from sympy import *
init_printing()
```

```python
x = symbols("x")
```

```python
y = x**2 + 2*x + 2
y
```

```python
solve(y)
```

```python
##  Reference for sympy.plot is https://docs.sympy.org/latest/modules/plotting.html?highlight=plot#sympy.plotting.plot.plot
plot(y)
```

```python
plot(y,(x,-5,5), ylim = (-3,3))  # simple added features
```
